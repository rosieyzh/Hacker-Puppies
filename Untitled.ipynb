```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
breeds = pd.read_csv("labels.csv")
test = pd.read_csv("sample_submission.csv")
```

```python
breeds
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelBinarizer
```

```python
encoder = LabelBinarizer()
one_hot = encoder.fit_transform(breeds['breed'])
one_hot.shape
```

```python
Y_train = np.asarray(one_hot)
```

```python
x_train = []
y_train = Y_train
x_test = []
```

```python
from keras.preprocessing import image
from keras.applications.resnet50 import ResNet50
from keras.applications.resnet50 import preprocess_input, decode_predictions
for f, x in breeds.values:
    Image = image.load_img("train/" + f + ".jpg", target_size=(224, 224))
    data = image.img_to_array(Image)
    data = preprocess_input(data)
    x_train.append(data)
```

```python
x_train
```

```python
labels = np.array(y_train, np.uint8) #data, datatype
features = np.array(x_train, np.float64)
```

```python
print(labels.shape)
print(features.shape)
```

```python
xtrain, xvalid, ytrain, yvalid = train_test_split(features, labels, test_size=0.25, random_state=1)
```

```python
from keras import Model
from keras.layers import Input, Dense

initial_model = ResNet50(weights=None)
a = initial_model.output
predictions = Dense(120, activation='relu')(a)

model = Model(inputs=initial_model.input, outputs=predictions)

model.compile(optimizer='rmsprop', loss='categorical_crossentropy')

model.summary()
```

```python
model.fit(xtrain, ytrain, batch_size=100, epochs=2, verbose=1)
```
