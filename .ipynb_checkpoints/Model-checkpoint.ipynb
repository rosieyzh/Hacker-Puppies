# Preprocessing

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

breeds = pd.read_csv("labels.csv")
breeds.head()
```

```python
import os, os.path
from PIL import Image

def load_image(file, directory) :
    file = file + ".jpg"
    imagesList = listdir(directory)
    image = Image.open(os.path.join(directory,file))
    image.load()
    image = image.resize((196,196))
    data = np.asarray(image, dtype="float64")
    return data
```

```python
training = []
for id in breeds['id']:
    training.append((id, load_image(id, 'train/')))
pd.DataFrame(training)
```
